````python
#1. Define a function named `initialize_inventory` that takes `products` as a parameter.
# Inside the function, implement the code for initializing the inventory dictionary using a loop and user input.

#List comprehension of function 1.
       
def initialize_inventory(products):
    inventory = {item: max(0, int(input(f"Enter the quantity of {item}s available: "))) for item in products}
            
    # Imprime los resultados finales
    for item, quantity in inventory.items() :
        if inventory[item] <= 0:
            print(f"Error: The quantity of {item} cannot be negative. The value has been set to 0.")
        print(f"Enter the number of {item} available: {inventory[item]} ")
            
            
    return inventory

#2.Define a fuction name "get_customers_orders". Implement the code for prompting the user to enter the product names using a loop.
#The fuction should return "customer_orders" set

def get_customer_orders(products,inventory) :

    num_orders = max(0,int(input("How many products would you like to order?")))
    customer_orders = set (input(f"Enter the name of a product that a customer wants to order, remember that it should be only these items {products}: ").strip() for x in range(num_orders))
    print(f"Enter the name of a product that a customer wants to order:{customer_orders}")
    return customer_orders


#3.Define a new fuction that calculates the total price of the customer order

def pricing_customer_order(customer_orders):
    
    total_price=0

    for order in customer_orders:
        price=max(0, float(input(f"Enter the price of {order}:")))
        print(f"\nEnter the price of {order}:{price}")
        total_price+=price
    print(f"\nThe total price of your customer order is {total_price:.2f}")
    return

#3. Define a fuction named "updated_inventory". Inside the fuction update the inventory dictionary based on customers orders

def updated_inventory(customer_orders,inventory) :


    #9.Update the inventory
    for item in customer_orders:
        if inventory[item]>0:
            inventory[item]-=1

        else:
            print(f"Cannot reduce {item}, no stock available!")

    inventory={item:quantity for item,quantity in inventory.items() if quantity > 0}
    
    return inventory


#4. Define a fuction named "calculated_orders_statistics". Implement code for calculating order statistics


def calculated_orders_statistics(customer_orders, products) :

    #7.1 Total products ordered
    total_products_ordered=len(customer_orders)
    total_available_products=len(products)
    print(f"\nTotal products ordered: {total_products_ordered}")


    #7.2 Percentage of products ordered
    percentage_ordered=(total_products_ordered/total_available_products)*100
    print(f"Percentage ordered compared to the total available products {percentage_ordered}%") 

    
    return total_products_ordered,percentage_ordered


#5. Define a fuction named "print_order_statistics". Implement code for printing

def print_order_statistics(order_statistics) :
    total_products_ordered, percentage_ordered = order_statistics
    print(f"\nOrder Statistics:\nTotal Products Ordered: {total_products_ordered}\nPercentage of Products Ordered: {percentage_ordered:.2f}%\n")
    

#6. Define a fuction named "print_updated_inventory". Implement code for printing

def print_updated_inventory(inventory) :

    #10.Print the Updated inventory
    print("\nUpdated inventory:")
    for item,quantity in inventory.items() :
        print(f"{item}:{quantity}")




#7. Called the fuctions.

#Define a list called products with different fashion clothes
products=["t-shirt","mug","hat","book","keychain"]

#Inicilize inventory
inventory=initialize_inventory(products)
#Obtein order
customer_orders=get_customer_orders(products,inventory)
#Calculate statistics
order_statistics=calculated_orders_statistics(customer_orders, products)
#Print statistics
print_order_statistics(order_statistics)
#Update inventory
inventory=updated_inventory(customer_orders,inventory)
#Print inventory
print_updated_inventory(inventory)
#Print total price
pricing_customer_order(customer_orders)
````
